```python
duplicates = df.duplicated()  
num_duplicates = duplicates.sum()

print('Sum of duplicate rows: ', num_duplicates)
```

```python
enrollment_threshold = 200
threshold = df[ df['Enrollment'] > enrollment_threshold ]
print('Enrollment greater than a 200: ',len(threshold))
```

```python
trails='No Results Available'

result=df[df['Study Results'] == trails]
print('Clinical trials have No Results Available: ',len(result))
```

```python
status1 = 'Completed'
status2 = 'Recruiting'
result = df[ (df['Status'] == status1) | (df['Status'] == status2) ]
num_trials = len(result)     
print(f"Number of trials in 'Completed' or 'Recruiting' status: {num_trials}")
```

```python
covid_trials = df[df['Acronym'] == 'COVID-19']
print('Clinical trials are related to COVID-19: ',len(covid_trials))
```

```python
threshold_date = 'January 1, 2020'

after_2020 = df[df['Start Date'] > threshold_date]
print('Clinical trials started after January 1, 2020: ', len(after_2020))
```
